```python
import csv

max_temp = 0 #최고 기온 저장 변수

with open("./datas/temp_seoul.csv") as f:
    data = csv.reader(f) #f- 파일 객체
    header = next(data) # 제목행 next()- 다음행으로 이동 삭제.
#     print(header)
    for row in data:
        if row[-1] =='': #최고기온이 누락된경우 ( 빈숫자열)
            row[-1] = -999 #터무니없는 숫자로 설정
        row[-1] = float(row[-1]) # 최고기온을 숫자형으로 변환(float()-실수형)
        row[0] = row[0].replace('\t','')
#         print(row)
#     최대값구하기
        if max_temp < row[-1]:
            max_temp = row[-1]
            max_date = row[0]
            
print(f'최고기온 : {max_temp}, 날짜 : {max_date}')
        
# 최고기운을 숫자형을 변환(float()-실수형)
```

# 최고 기온 데이터 리스트를 시각화



```python
#최고 기온 데이터를 시각화

import csv
import matplotlib.pyplot as plt

max_temp = 0 # 최고 기온 저장 변수
max_date = 0

with open("./datas/temp_seoul.csv") as f:
    data = csv.reader(f)   # f- 파일 객체 
    header = next(data)  # 제목행 next() - 다음행으로 이동
    result = [] # 최고 기온ㄱ밧들을 저장할 리스트
    # print(header)
    for row in data:
        # 최고기온을 숫자형으로 변환(float()-실수형)
        if row[-1] != '': #최고기온이 누락된 경우(빈 문자열)
            result.append(float(row[-1]))
            
#print(len(result))
#print(result)
plt.plot(result,'r')
plt.show()
```

## 내 생일에 최고기온을 시각화하기

```python
with open("./datas/temp_seoul.csv") as f:
    data = csv.reader(f)
    next(data)
    result = [] #최고 기온값저장
    for row in data:
        #생일이 5월 11이라면
        if row[0] != "":#초괴기온이 빈 문자열이 아니면
        if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '11': #[wt,1907,10,01]
                result.append(float(row[-1]))
plt.plot(result)
```

```python
with open("./datas/temp_seoul.csv") as f:
    data = csv.reader(f) #data가 리스틀 저장
    next(data)
    high = []
    low = []
   
    for row in data:
        #생일이 5월 11이라면
        if row[-1] != "" and row[-2]:#최고기온이 빈 문자열이 아니면
            if int(row[0].split('-')[0])>1980: #연도가 1980년 보다 크면 (정수형)
                if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '22': #[wt,1907,10,01]
                    
                    high.append(float(row[-1])) #실수로 변환
                    low.append(float(row[-2]))
plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus']
plt.title("내 생일의 기온 변화 그래프")
plt.plot(high, 'hotpink')
plt.plot(low, 'skyblue')
plt.xlabel('연도')
plt.ylabel('기온')
plt.show()
```
